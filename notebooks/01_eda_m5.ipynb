```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
calendar = pd.read_csv("../data/raw/calendar.csv")
sales = pd.read_csv("../data/raw/sales_train_validation.csv")
sell_prices = pd.read_csv("../data/raw/sell_prices.csv")

# Quick info
print("Calendar:")
print(calendar.head())
print("Sales:")
print(sales.head())
print("Sell Prices:")
print(sell_prices.head())

# Check shape and missing values
print("\nShapes:", calendar.shape, sales.shape, sell_prices.shape)
print("\nMissing values:")
print(calendar.isna().sum(), "\n", sales.isna().sum().sum(), "\n", sell_prices.isna().sum())

# Example plot for one product
sample_product = sales.iloc[0, 6:]  # first product's daily sales
plt.figure(figsize=(12,4))
plt.plot(sample_product.values)
plt.title("Sample Product Sales Over Time")
plt.show()
```
